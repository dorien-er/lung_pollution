```python
import pandas as pd

covid_data = pd.read_csv('/Users/saraisidebroggini/code/dorien-er/lung_pollution/lung_pollution/data/covid_pollution_clean.csv')

covid_data.shape
covid_data = covid_data[['BL','county','EWZ','Shape__Area', 'death_rate', 'cases', 'deaths','cases_per_100k']]
covid_data.head()
covid_data['deaths_per_100k'] = covid_data['deaths']/covid_data['EWZ']*100000
covid_data.head()
berlin = covid_data[covid_data["BL"] == 'Berlin']
covid_data['county'][399] = 'Berlin'
covid_data.loc[399] 
berlin_sum = berlin[['Shape__Area', 'cases', 'deaths']].sum()
berlin_average = berlin[['death_rate','cases_per_100k', 'deaths_per_100k']].mean()
covid_data['cases'][399] = berlin_sum.cases
covid_data['Shape__Area'][399] = berlin_sum.Shape__Area
covid_data['deaths'][399] = berlin_sum.deaths
covid_data['death_rate'][399] = berlin_average.death_rate
covid_data['cases_per_100k'][399] = berlin_average.cases_per_100k
covid_data['deaths_per_100k'][399] = berlin_average.deaths_per_100k
covid_data.drop(index=[400,401,402,403,404,405,406,407,408,409,410], axis=0, inplace=True)
```

```python
url = 'https://raw.githubusercontent.com/isellsoap/deutschlandGeoJSON/main/4_kreise/1_sehr_hoch.geo.json'
dfObj = pd.read_json(url)

list_ids = []
list_counties = []
list_coords = []
list_code_counties = []
for i in range(len(dfObj['features'])):
    list_ids.append(dfObj["features"][i]["id"])
    list_counties.append(dfObj["features"][i]["properties"]["NAME_3"])
    list_coords.append(dfObj["features"][i]["geometry"]['coordinates'][0])
    list_code_counties.append(dfObj["features"][i]["properties"]["ID_3"])
    

df = pd.DataFrame(list(zip(list_ids, list_counties, list_coords,list_code_counties)), columns =['Id', 'County', 'Coordinates','Id_counties'])
```
