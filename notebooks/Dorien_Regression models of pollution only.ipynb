# Load data and keep 2019 timepoint

```python
import pandas as pd

df_covid_AP = pd.read_csv('../lung_pollution/data/covid_pollution_clean02.csv')

df_covid_AP.drop(columns=['Unnamed: 0'], inplace = True)
df_covid_AP = df_covid_AP.rename(columns={"PM2.5_annualMean": "PM2_5_annualMean", "Fully vaccinated": "Fully_vaccinated", "Population density": "Population_density"})
```

# Feature engineering

```python
df_num = df_covid_AP_2019[['NO2_annualMean', 'NO_annualMean','O3_daily8HrMax','PM2_5_annualMean', 'cases_per_100k', 'deaths_per_100k']]
df_num_2018 = df_covid_AP_2018[['NO2_annualMean', 'NO_annualMean','O3_daily8HrMax','PM2_5_annualMean', 'cases_per_100k', 'deaths_per_100k']]
```

```python
X_train = df_covid_AP_2019[['NO2_annualMean', 'NO_annualMean', 'O3_daily8HrMax','PM2_5_annualMean']]
X_train[['lnNO2']] = np.log(X_train[['NO2_annualMean']])
X_train[['lnNO']] = np.log(X_train[['NO_annualMean']])
X_train[['lnO3']] = np.log(X_train[['O3_daily8HrMax']])
X_train[['lnPM2_5']] = np.log(X_train[['PM2_5_annualMean']])
y_train = df_covid_AP_2019[['cases_per_100k']]
```

```python
X_train.insert(8, "dNO2", df_num_2018['NO2_annualMean']-df_num['NO2_annualMean'])
X_train.insert(9, "dNO", df_num_2018['NO_annualMean']-df_num['NO_annualMean'])
X_train.insert(10, "dO3", df_num_2018['O3_daily8HrMax']-df_num['O3_daily8HrMax'])
X_train.insert(11, "dPM2_5", df_num_2018['PM2_5_annualMean']-df_num['PM2_5_annualMean'])
```

```python
X_train.drop(columns='lnNO', inplace=True)
```
